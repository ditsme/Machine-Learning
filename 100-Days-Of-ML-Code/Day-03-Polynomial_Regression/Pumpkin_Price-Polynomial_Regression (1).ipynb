```python
import numpy as np
import pandas as pd

data=pd.read_csv("pumpkin.csv")
data.head(5)
```

```python
data.drop(['Type','Sub Variety','Grade','Environment','Unit of Sale','Quality','Condition','Appearance','Storage','Crop','Repack','Trans Mode'],axis=1,inplace=True)
```

```python
data.head(5)
```

```python
data.drop(['City Name','Package','Variety','Origin','Origin District','Color'],axis=1,inplace=True)
```

```python
data.head(5)
```

```python
cat_map = {
    'sml': 0,
    'med': 1,
    'med-lge': 2,
    'lge': 3,
    'xlge': 4,
    'exjbo': 5
}
data = data.assign(
    size=data['Item Size'].map(cat_map),
    price=data['High Price'] + data['Low Price'] / 2,
    size_class=(data['Item Size'].map(cat_map) >= 2).astype(int)
)
```

```python
data.head(5)
```

```python
data.drop(['Commodity Name','Date','Low Price','High Price','Mostly Low','Mostly High','Item Size'],axis=1,inplace=True)
```

```python
data.head(5)
```

```python
data.dropna()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
y = data.values[:,1]
x = data.values[:,[0,2]]
x_test=[[2.0,1]]
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# PolynomialFeatures (prepreprocessing)
poly = PolynomialFeatures(degree=2)
X = poly.fit_transform(x)
X_test = poly.fit_transform(x_test)
```

```python
# Instantiate
lg = LinearRegression()

# Fit
lg.fit(X, y)

# Obtain coefficients
lg.coef_
```

```python
# Predict
lg.predict(X_test)
```
